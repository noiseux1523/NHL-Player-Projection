# Reformat CapFriendly Data

```python
import pandas as pd
import glob
```

## 1. Load Data

```python
data = []
for file in glob.glob('../data/CapData*.csv'):
    year = int(file[-8:-4])
    year_data = pd.read_csv(file)
    year_data['PLAYER'] = year_data['PLAYER'].apply(lambda x: x.split('.')[-1].strip())
    year_data['YEAR'] = year
    year_data['CAP HIT'] = year_data['CAP HIT'].str.replace(r'\$', '').replace(',', '', regex=True).astype(float)
    data.append(year_data)
data = pd.concat(data).reset_index(drop=True)
```

```python
# Contracts per year
data.groupby('YEAR').size()
```

```python
# Check for players with current and incoming new contract already signed
new_contract = data.groupby('PLAYER')['CAP HIT'].apply(set).reset_index()
new_contract[new_contract['CAP HIT'].apply(lambda x: len(x))>1]
```

```python
data = data.sort_values(['YEAR', 'PLAYER'], ascending=False).drop_duplicates(subset='PLAYER', keep='first')
data.head()
```

```python
data.to_csv('../data/capfriendly_data.csv', index=False)
```
